```python
cols_to_drop = ['alive', 'alone']
df = df.drop(columns=cols_to_drop)
df.columns
```

```python
avg_age = round(df['age'].mean())
print(avg_age)
```

```python
df['age']= df['age'].fillna(avg_age) #fillna command replaces
# the null values in a particular labels by your given numbers or varliables
df.isnull().sum()
```

```python
# dropna drops null values
df = df.dropna(subset =['embarked'])
df.isnull().sum()
```
