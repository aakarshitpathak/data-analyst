- row wise selection 
    - iloc - index based selection
    - loc - labes based selection
- column wise selection
    - single column selection
        - dictionary like selection
        - object column selection
    
    - multiple column selection
        - dictionary like selection

```python
import pandas as pd

df = pd.read_csv("kc_house_data.csv")
df
df.iloc[2]
# Note: we use iloc when no unique id is given
```

```python
df.iloc[21612]
```

```python
df.iloc[3:9]
```

```python
df.iloc[3:9:2]
```

```python
df.iloc[[3,9,11,17,19]] 
# Note: IF YOU ARE DOING SLICING FOR PARTICULAR DATA THEN 
# DOUBLE BRACKETTS ARE NOT NECESSARY BUT IF YOU ARE GOING FOR
# MULTIPLE DATA THE DOUBLE BRACKETTS ARE NECESSARY
```

```python
df.iloc[[2,4,7],[3,5,7]] #select specific rows and column
```
