```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set plots for graph
fig,ax= plt.subplots()

# set background color
fig.set_facecolor('lightgreen')
ax.set_facecolor('lightblue')

# set title,x-axis and y-axis
ax.set_title('Line Chart',color = 'brown', fontsize=15 )
ax.set_xlabel('X-Axis', color = 'darkblue', fontsize=12)
ax.set_ylabel('Y-Axis', color = 'darkgreen', fontsize=12)

# plotting the coordinates of x and y axis
x=[12,14,16,11,7,9,18,20,17,4]
y=[102,69,189,182,200,169,45,89,144,192]
ax.plot(x,y,color='red',
linestyle='-',
linewidth=2,
marker='*',
markersize=8,
markeredgecolor='pink',
markerfacecolor='black',
markeredgewidth=1,
) 


```
