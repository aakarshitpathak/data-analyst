```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
# Age distribution
fig = sns.histplot(df, x ='age' , color ="green" , kde=True)
# kde=kernel density estimate
fig.set_title('Age Distribution of Titanic Passengers')
fig.set_xlabel('Age')
fig.set_ylabel('Count')
```

```python
# survival count vased on gender
fig = sns.countplot(df, x='alive',hue='sex')
fig.set_title('Survival count of titanic passengers')
fig.set_xlabel ('Survived')
fig.set_ylabel('Count')
```

```python
# average fare by passenger class and gender
fig=sns.barplot(df, x='class',y='fare', hue='sex',palette='viridis')
fig.set_title('Average fare by passenger and class')
fig.set_xlabel('Passenger Class')
fig.set_ylabel('Average Fare')
```

```python
import seaborn as sns
# average fare by passenger gender
fig=sns.barplot(df, x='sex',y='fare')
fig.set_title('Average fare by gender')
fig.set_xlabel('Passenger gender')
fig.set_ylabel('Average Fare')
```

```python
# passenger count by class
fig = sns.countplot(df, x='class' , color = "Yellow")
fig.set_title('Class count of titanic passengers')
fig.set_xlabel ('Class')
fig.set_ylabel('Count')
```
