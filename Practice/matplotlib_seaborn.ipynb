```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.get_dataset_names()
df
```

```python
df=sns.load_dataset('flights')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter only 1949 data
flights_1949 = df[df['year'] == 1949]

# Use barplot instead of countplot
fig = sns.barplot(data=flights_1949, x='month', y='passengers')

# Titles and labels
fig.set_title("Total Passengers Traveled in Each Month (1949)")
fig.set_xlabel("Months")
fig.set_ylabel("Number of Passengers")

plt.show()
```

```python
# year wise total number of passengers travelled in flights

# group by year and sum passengers
yearly_pass=df.groupby('year')['passengers'].sum().reset_index()
print(yearly_pass) # to check the total 

fig=sns.histplot(data=yearly_pass,x='year',y='passengers')
fig.set_title("Passengers count by year")
fig.set_xlabel("Year")
fig.set_ylabel("Passengers count")

plt.show()
```

```python
# year wise total number of passengers travelled in flights
# but by using bar plot 

yearly_pass=df.groupby('year')['passengers'].sum().reset_index()
print(yearly_pass)

# plot the bar m
fig=sns.barplot(data=yearly_pass,x='year',y='passengers')
fig.set_title("Passengers count by year")
fig.set_xlabel("Year")
fig.set_ylabel("Passengers count")

plt.show()
```
